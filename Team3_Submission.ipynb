## ***PPDB +SARI MODEL***     
*Anushka Panda*


*This model will try to simplify Wikipedia articles using paraphrasing , instead of simply deleting seemingly unimportant tough content. Also, unlike most models , which compare outputs to only human reference sentences while training , this model shall also try and compare the outputs to the input itself.*

*SARI : System output Against References and Input sentences . 
Let input sentence be fed into the model . Some human references shall also be fed in as a part of the model , and based on them, the model will generate an output sentence.*

*Two functions : Precision and Recall*

*Three operations to simplify : add, keep, delete a particular n - gram , with the help of the human references .*

*For adding and keeping, both precision and recall is calculated for each n gram, and finally the average over the n grams is calculated, and the value of a function which incorporates both precision and recall, is calculated. For deleting, only precision is calculated.*

*The SARI score is the mean of the functions from adding and keeping, and the precision of deleting. A high SARI score implies more simplicity.
The model calculates SARI scores for all possible simplifications of the input, and returns the output with the highest SARI score.*

*For paraphrasing, a vector containing some functions according to the input and output sentences is dotted with a weight matrix which represents all rules of paraphrasing, and that gives the final simplification. The matrix can be trained by comparing the output with a bad output.
The cycle repeats for every sentence.*

## ***SUMMARY***
*Shreyasi Mandal*

*We examine the task of sentence simplification which aims to reduce the reading complexity of a sentence by incorporating more accessible vocabulary and sentence structure. We introduce a new data set that pairs English Wikipedia with Simple English Wikipedia and is orders of magnitude larger than any previously examined for sentence simplification. The data contains the full range of simplification operations including rewording, reordering, insertion and deletion. We provide an analysis of this corpus as well as preliminary results using a phrase-based translation approach for simplification.*

*The task of text simplification aims to reduce the complexity of text while maintaining the content. We explore the sentence simplification problem: given a sentence, the goal is to produce an equivalent sentence where the vocabulary and sentence structure are simpler.*

*It is widely accepted that sentence simplification involves three different elements: splitting, deletion and paraphrasing.*

*As a preprocessing step, simplification can improve the performance of NLP tasks, including parsing, semantic role labeling, machine translation and summarization.*

*Finally, models for text simplification are similar to models for sentence compression; advances in simplification can benefit compression, which has applications in mobile devices, summarization and captioning.*
